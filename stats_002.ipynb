```python
from scipy import stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
data_normal = stats.norm.rvs(size = 1000,
                             loc = 0,
                             scale = 1)
```

```python
sns.displot(data_normal,
            bins = 100)
plt.show()
```

```python
data_uniform = stats.uniform.rvs(size = 1000,
                                 loc = 0,
                                 scale = 1)
```

```python
sns.displot(data_uniform,
            bins = 100)
plt.show()
```

```python
data_bernouli = stats.bernoulli.rvs(size = 1000,
                                    p = 0.2)
```

```python
sns.displot(data_bernouli,
            bins = 100)
```

```python
data_binom = stats.binom.rvs(size = 1000,
                             n = 100,
                             p = 0.2)
```

```python
data_binom.std()
```

```python
sns.displot(data_binom,
            bins = 100)
```

```python
data_multinomial = np.random.multinomial(size = 1000,
                                         n = 6,
                                         pvals = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
data_multinomial
```

```python
sns.set_context('paper',
                font_scale=1.4)



#sns.displot(data_multinomial,
 #           bins = 100)

plt.hist(data_multinomial,rwidth=1, bins=6,histtype='bar')
plt.legend()
plt.show()
```

```python
data_mulinomial2 = stats.multinomial.rvs(size=1000,
                                         n=6,
                                         p=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

plt.hist(data_mulinomial2,rwidth=1, bins=6,histtype='bar')

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
